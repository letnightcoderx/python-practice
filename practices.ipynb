```python
# Reverse digits of a number using while loop
num = int(input("Enter a positive integer: "))

while num <= 0:
    print("Please enter a positive integer!")
    num = int(input("Enter a positive integer: "))

original = num

reversed_num = 0

while num > 0:
    digit = num % 10           # Get last digit
    reversed_num = reversed_num * 10 + digit  # Add digit to reversed number
    num = num // 10            # Remove last digit

print(f"The reverse of {original} is: {reversed_num}")
```

```python
# Palindrome number checker (with input validation)

# Ask the user for a positive integer
num = int(input("Enter a positive integer: "))

# Validate input - keep asking until positive
while num < 0:
    print("Please enter a positive integer!")
    num = int(input("Enter a positive integer: "))

# Save the original number for comparison
original = num

# Initialize reversed number
reversed_num = 0

# Reverse the number using while loop
while num > 0:
    digit = num % 10                    # Get last digit
    reversed_num = reversed_num * 10 + digit  # Build reversed number
    num = num // 10                     # Remove last digit

# Compare original with reversed
if original == reversed_num:
    print("It is a Palindrome.")
else:
    print("It is NOT a Palindrome.")
```

```python
# Armstrong number checker

# Ask the user for a positive integer
num = int(input("Enter a positive integer: "))

# Validate input - keep asking until positive
while num < 0:
    print("Please enter a positive integer!")
    num = int(input("Enter a positive integer: "))

# Save the original number for comparison
original = num

# Step 1: Count the number of digits
count = 0
temp = num  # Use temp to preserve num for later use

# Handle special case: 0 has 1 digit
if temp == 0:
    count = 1
else:
    while temp > 0:
        temp = temp // 10
        count = count + 1

# Step 2: Calculate sum of digits raised to power of count
temp = num  # Reset temp to original number
sum_of_powers = 0

while temp > 0:
    digit = temp % 10                    # Get last digit
    sum_of_powers = sum_of_powers + (digit ** count)  # Add digit^count
    temp = temp // 10                    # Remove last digit

# Step 3: Check if Armstrong number
if original == sum_of_powers:
    print(f"{original} is an Armstrong number.")
else:
    print(f"{original} is NOT an Armstrong number.")
```

```python
# Prime number checker

# Ask the user for a positive integer
num = int(input("Enter a positive integer: "))

# Validate input - keep asking until positive
while num <= 0:
    print("Please enter a positive integer!")
    num = int(input("Enter a positive integer: "))

# Handle special cases: 0 and 1 are not prime
if num == 0 or num == 1:
    print(f"{num} is NOT a prime number.")
else:
    # Assume the number is prime initially
    is_prime = True
    
    # Start checking divisors from 2
    divisor = 2
    
    # Check divisibility up to num // 2 (or we could use divisor * divisor <= num)
    while divisor * divisor <= num:
        if num % divisor == 0:
            is_prime = False  # Found a divisor, not prime
            break
        divisor += 1
    
    # Print the result
    if is_prime:
        print(f"{num} is a prime number.")
    else:
        print(f"{num} is NOT a prime number.")
```

```python
# 🏗️ Mini Project 1: Number Utility Program

# Main menu loop
while True:
    print("\n===== Number Utility =====")
    print("1. Check Even or Odd")
    print("2. Check Prime")
    print("3. Check Palindrome")
    print("4. Reverse Number")
    print("5. Count Digits")
    print("6. Exit")
    
    # Get user choice
    choice = int(input("\nEnter your choice: "))
    
    # Exit option
    if choice == 6:
        print("Thank you for using Number Utility. Goodbye!")
        break
    
    # Validate choice
    if choice < 1 or choice > 6:
        print("Invalid choice! Please enter a number between 1 and 6.")
        continue
    
    # Ask for a number
    num = int(input("\nEnter a number: "))
    
    # Validate number
    while num < 0:
        print("Please enter a positive integer!")
        num = int(input("Enter a number: "))
    
    # Perform operation based on choice
    if choice == 1:
        # Check Even or Odd
        if num % 2 == 0:
            print(f"\n{num} is an Even number.")
        else:
            print(f"\n{num} is an Odd number.")
    
    elif choice == 2:
        # Check Prime
        if num == 0 or num == 1:
            print(f"\n{num} is NOT a prime number.")
        else:
            is_prime = True
            divisor = 2
            while divisor * divisor <= num:
                if num % divisor == 0:
                    is_prime = False
                    break
                divisor += 1
            
            if is_prime:
                print(f"\n{num} is a Prime number.")
            else:
                print(f"\n{num} is NOT a prime number.")
    
    elif choice == 3:
        # Check Palindrome
        original = num
        reversed_num = 0
        temp = num
        
        while temp > 0:
            digit = temp % 10
            reversed_num = reversed_num * 10 + digit
            temp = temp // 10
        
        if original == reversed_num:
            print(f"\n{original} is a Palindrome.")
        else:
            print(f"\n{original} is NOT a Palindrome.")
    
    elif choice == 4:
        # Reverse Number
        original = num
        reversed_num = 0
        
        while num > 0:
            digit = num % 10
            reversed_num = reversed_num * 10 + digit
            num = num // 10
        
        print(f"\nThe reverse of {original} is: {reversed_num}")
    
    elif choice == 5:
        # Count Digits
        if num == 0:
            count = 1
        else:
            count = 0
            temp = num
            while temp > 0:
                temp = temp // 10
                count += 1
        
        print(f"\n{num} has {count} digit(s).")
```

```python
# Create the list
numbers = [12, 45, 7, 89, 23, 56]

# Initialize largest and second_largest
largest = numbers[0]
second_largest = None

for num in numbers[1:]:
    if num > largest:
        # Current largest becomes second largest
        second_largest = largest
        largest = num
    elif num != largest:
        # Update second_largest if num is between second_largest and largest
        if second_largest is None or num > second_largest:
            second_largest = num

# Print the result
print("Second largest number:", second_largest)
```

```python
# Function definition
def calculate(a, b):
    addition = a + b
    subtraction = a - b
    multiplication = a * b

    if b == 0:
        division = "Cannot divide by zero"
    else:
        division = a / b

    return addition, subtraction, multiplication, division

# Ask the user for input
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

# Store the returned values
add, sub, mul, div = calculate(num1, num2)

# Print the results
print("Addition:", add)
print("Subtraction:", sub)
print("Multiplication:", mul)
print("Division:", div)
```

```python
# 🏗️ Mini Project #2: Function to calculate total marks
def get_total(marks):
    return sum(marks)

# Function to calculate average
def get_average(total, subjects):
    return total / subjects

# Function to determine grade
def get_grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 80:
        return "B"
    elif avg >= 70:
        return "C"
    elif avg >= 60:
        return "D"
    else:
        return "F"

# Main Program

marks = []

# Take input for 5 subjects
for i in range(1, 6):
    mark = float(input(f"Enter marks for Subject {i}: "))
    marks.append(mark)

# Call functions
total = get_total(marks)
average = get_average(total, len(marks))
grade = get_grade(average)

# Print report
print("\n------ Student Report ------")
print("Marks:", marks)
print("Total:", total)
print("Average:", average)
print("Grade:", grade)
```
